### using neural networks (keras+theano)

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from keras.layers.core import Lambda

def glm_poisson_keras(Xr, Yr, Xt):
    
    model = Sequential()
    model.add(Dense(1, input_dim=np.shape(Xr)[1], init='uniform', activation='linear'))
    model.add(Lambda(lambda x: np.exp(x)))
    model.compile(loss='poisson', optimizer='rmsprop')

    model.fit(Xr, Yr, nb_epoch=3, batch_size=16, verbose=0, validation_split=0.0)
    
    Yt = model.predict_proba(Xt, verbose=0)
    
    return Yt[:,0]
```

```python
Yt_hat_glm_keras = glm_poisson_keras(X, y, Xangles)
```

### let's compute cv scores

#### first define some functions

```python
from sklearn.cross_validation import KFold
from sklearn.cross_validation import LabelKFold

#---------------------------------------
def glm_poisson_pyglmnet(Xr, Yr, Xt):
    glm = GLM(distr='poissonexp', alpha=0., reg_lambda=[0.], tol=1e-6)
    glm.fit(Xr, Yr)
    Yt = glm.predict(Xt)[0]
    return Yt

def poisson_pseudoR2(y, yhat, ynull):
    eps = np.spacing(1)
    L1 = np.sum(y*np.log(eps+yhat) - yhat)
    L1_v = y*np.log(eps+yhat) - yhat
    L0 = np.sum(y*np.log(eps+ynull) - ynull)
    LS = np.sum(y*np.log(eps+y) - y)
    R2 = 1-(LS-L1)/(LS-L0)
    return R2

def fit_cv(X, Y, algorithm = 'XGB_poisson', n_cv=10, verbose=1, label=[]):

    if np.ndim(X)==1:
        X = np.transpose(np.atleast_2d(X))

    if len(label)>0:
        skf  = LabelKFold(np.squeeze(label), n_folds=n_cv)
    else:
        skf  = KFold(n=np.size(Y), n_folds=n_cv, shuffle=True, random_state=42)

    i=1
    Y_hat=np.zeros(len(Y))
    pR2_cv = list()
    for idx_r, idx_t in skf:
        if verbose > 1:
            print '...runnning cv-fold', i, 'of', n_cv
        i+=1
        Xr = X[idx_r, :]
        Yr = Y[idx_r]
        Xt = X[idx_t, :]
        Yt = Y[idx_t]

        Yt_hat = eval(algorithm)(Xr, Yr, Xt)
        Y_hat[idx_t] = Yt_hat

        pR2 = poisson_pseudoR2(Yt, Yt_hat, np.mean(Yr))
        pR2_cv.append(pR2)

        if verbose > 1:
            print 'pR2: ', pR2

    if verbose > 0:
        print("pR2_cv: %0.6f (+/- %0.6f)" % (np.mean(pR2_cv),
                                             np.std(pR2_cv)/np.sqrt(n_cv)))

    return Y_hat, pR2_cv

def plot_model_comparison(models_for_plot, models=[], color='r', title=None):

    plt.plot([-1, len(models_for_plot)], [0,0],'--k', alpha=0.4)

    mean_pR2 = list()
    sem_pR2 = list()

    for model in models_for_plot:
        PR2_art = models[model]['PR2']
        mean_pR2.append(np.mean(PR2_art))
        sem_pR2.append(np.std(PR2_art)/np.sqrt(np.size(PR2_art)))

    plt.bar(np.arange(np.size(mean_pR2)), mean_pR2, 0.8, align='center',
            ecolor='k', alpha=0.3, color=color, ec='w', yerr=np.array(sem_pR2),
            tick_label=models_for_plot)
    plt.plot(np.arange(np.size(mean_pR2)), mean_pR2, 'k.', markersize=15)

    plt.ylabel('pseudo-R2')
    ax = plt.gca()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.tick_params(axis='y', right='off')
    plt.tick_params(axis='x', top='off')
    if title:
        plt.title(title)
```

```python
Models = dict()
```

```python
X = list()
X.append(np.cos(dir_vel))
X.append(np.sin(dir_vel))
X = np.transpose(np.array(X))

Models['glm_poisson_pyglmnet'] = dict()

Yt_hat, PR2 = fit_cv(X, y, algorithm = 'glm_poisson_pyglmnet', n_cv=10, verbose=2, label=[])
Models['glm_poisson_pyglmnet']['Yt_hat'] = Yt_hat
Models['glm_poisson_pyglmnet']['PR2'] = PR2
```

```python
X = list()
X.append(np.cos(dir_vel))
X.append(np.sin(dir_vel))
X = np.transpose(np.array(X))

Models['glm_poisson_keras'] = dict()

Yt_hat, PR2 = fit_cv(X, y, algorithm = 'glm_poisson_keras', n_cv=10, verbose=2, label=[])
Models['glm_poisson_keras']['Yt_hat'] = Yt_hat
Models['glm_poisson_keras']['PR2'] = PR2
```

### it sucks, it needs engineered features, let's add some hidden layers to the NN

```python
def NN_poisson(Xr,Yr, Xt):

    
    if np.ndim(Xr)==1:
        Xr = np.transpose(np.atleast_2d(Xr))
    
    model = Sequential()
    model.add(Dense(3000, input_dim=np.shape(Xr)[1], init='uniform', activation='tanh'))
    model.add(Dropout(0.6))
    model.add(Dense(100, init='uniform', activation='tanh'))
    model.add(Dropout(0.6))

    model.add(Dense(1, activation='linear'))
   
    model.add(Lambda(lambda x: np.exp(x)))
    
    model.compile(loss='poisson', optimizer='adam')

    hist = model.fit(Xr, Yr, nb_epoch=3, batch_size=32, verbose=1, validation_split=0.0)
    
    result = model.predict_proba(Xt)
    return result[:,0]
```

```python
X = list()
X.append(dir_vel)
X = np.transpose(np.array(X))

Models['NN_poisson_angle'] = dict()

Yt_hat, PR2 = fit_cv(X, y, algorithm = 'NN_poisson', n_cv=10, verbose=2, label=[])
Models['NN_poisson_angle']['Yt_hat'] = Yt_hat
Models['NN_poisson_angle']['PR2'] = PR2
```

### xgboost

```python
import xgboost as xgb
def XGB_poisson(Xr, Yr, Xt):
    param = {'objective': "count:poisson",
    'eval_metric': "logloss",
    'num_parallel_tree': 2,
    'eta': 0.07,
    'gamma': 1, # default = 0
    'max_depth': 1,
    'subsample': 0.5,
    'seed': 2925,
    'silent': 1,
    'missing': '-999.0'}
    param['nthread'] = 6

    dtrain = xgb.DMatrix(Xr, label=Yr)
    dtest = xgb.DMatrix(Xt)

    num_round = 200
    bst = xgb.train(param, dtrain, num_round)

    Yt = bst.predict(dtest)
    return Yt
```

```python
Models['xgb_poisson_angle'] = dict()
Yt_hat, PR2 = fit_cv(X, y, algorithm = 'XGB_poisson', n_cv=10, verbose=2, label=[])
Models['xgb_poisson_angle']['Yt_hat'] = Yt_hat
Models['xgb_poisson_angle']['PR2'] = PR2
```
